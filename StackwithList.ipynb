```python
# -*- coding: utf-8 -*-
"""
Created on Thu Jan 30 13:03:21 2020

@author: kisku
"""

class ArrayStack:
    top=-1
    def __init__(self,cap):
        self.capacity=cap
        self.myarray = []
    def isFull(self):
        if self.top==self.capacity-1:
            return(1)
        else:
            return(0)
    def isEmpty(self):
        if self.top==-1:
            return(1)
        else:
            return(0)
    def pushStack(self):
        if (self.isFull()==0):
            i=int(input("Enter a number: "))
            self.myarray.append(i)
            self.top+=1
        else:
            print("Stack is Full.")
    def popStack(self):
        if (self.isEmpty()==0):
            print(self.myarray[self.top])
            self.top-=1
        else:
            print("Stack is Empty.")
        
Stack = ArrayStack(5)

a=1
while(a==1):
    print("\n1. Push:")
    print("\n2. Pull:")
    print("\n3. Exit:")
    choice=int(input())
    if choice==1:
        Stack.pushStack()
    if choice==2:
        Stack.popStack()
    if choice==3:
        break
```
