```python
# df.query('a > b')
# df[df.a > df.b]  # same result as the previous expression

ks = ks.query("state!='live'") # removing entries for live
```

```python
ks.groupby('state')['ID'].count()
```

```python
'''assign(**kwargs) method of pandas.core.frame.DataFrame instance
    Assign new columns to a DataFrame, returning a new object
    (a copy) with the new columns added to the original ones.
    Existing columns that are re-assigned will be overwritten.'''

# Add outcome column, "successful" == 1, others are 0

ks = ks.assign(outcome=(ks.state=='successful').astype(int))
```

```python
# Converting the launched feature into categorical features
ks = ks.assign(hour=ks.launched.dt.hour,
              day=ks.launched.dt.day,
              month=ks.launched.dt.month,
              year=ks.launched.dt.year)
```

```python
# Assigning an integer to each value of the categorical feature and replaces those values with the integers.
# category, currency, and country

from sklearn.preprocessing import LabelEncoder

cat_features = ['category', 'currency', 'country']
encoder = LabelEncoder()
```

```python
encoded = ks[cat_features].apply(encoder.fit_transform)
```

```python
data = ks[['goal', 'hour', 'day', 'month', 'year', 'outcome']]
```

```python
data.head(10)
```

```python
data = data.join(encoded)
```
